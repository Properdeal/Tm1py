```python
from TM1py.Services import TM1Service

# TM1 Configuration
TM1_CONFIG = {
    "address": "YOUR_SERVER",  # e.g., "localhost"
    "port": 12354,            # REST API port (default is often 12354)
    "user": "YOUR_USERNAME",
    "password": "YOUR_PASSWORD",
    "ssl": True               # Set to False if SSL is not used
}

CUBE_NAME = "YOUR_CUBE_NAME"  # Replace with your cube name
DIMENSIONS = ["Dim1", "Dim2", "Dim3"]  # Replace with your cube dimensions
ELEMENTS = ["Value1", "Value2", "Value3"]  # Replace with specific dimension elements

def get_cube_data(tm1: TM1Service):
    """
    Retrieve data from the TM1 cube.
    """
    mdx_query = f"""
    SELECT
    {{[{DIMENSIONS[0]}].Members}} ON ROWS,
    {{[{DIMENSIONS[1]}].Members}} ON COLUMNS
    FROM [{CUBE_NAME}]
    WHERE ([{DIMENSIONS[2]}].[{ELEMENTS[2]}])
    """
    cellset = tm1.cells.execute_mdx(mdx_query)
    print("Cube Data Retrieved:")
    for coordinates, value in cellset.items():
        print(f"{coordinates}: {value}")


def feed_data_to_cube(tm1: TM1Service):
    """
    Feed data into the TM1 cube.
    """
    coordinates = (ELEMENTS[0], ELEMENTS[1], ELEMENTS[2])
    value = 100  # Value to write
    tm1.cubes.cells.write_value(value, CUBE_NAME, coordinates)
    print(f"Fed value {value} to {coordinates} in cube {CUBE_NAME}")


if __name__ == "__main__":
    with TM1Service(**TM1_CONFIG) as tm1:
        # Retrieve cube data
        get_cube_data(tm1)
        
        # Feed data into the cube
        feed_data_to_cube(tm1)
```
